```python
import pandas as pd
import numpy as np
import math
```

```python
data = pd.read_excel("udacity_inferential_stats.xlsx", header=None,names={"Klout"})
data
```

```python
mean_pop =data.mean()
mean_pop
```

```python
pop_std = data.std()
pop_std 
```

```python
data_sample_size_sqrt=math.sqrt(250)
data_sample_size_sqrt
```

```python
standard_error = pop_std/data_sample_size_sqrt
standard_error
```

# Z-score for mean 40 from sample size 250

```python
group_score=40
```

```python
(group_score-mean_pop)/standard_error
```

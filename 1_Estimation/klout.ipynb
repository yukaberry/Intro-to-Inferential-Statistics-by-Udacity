```python
data = pd.read_excel("udacity_inferential_stats.xlsx", header=None,names={"Klout"})
data
```

# What is the mean of this population?

```python
mean_pop =data.mean()
mean_pop
```

# What is the population standard deviation?

```python
pop_std = data.std()
pop_std 
```

# What is the mean of this distribution? 
# A. the same as the population mean

# What is the standard error? 


# A. Divide the population standard deviation by the square root of "n". "n" is the sample size. in this example, the sample size is 35 (not 1048)

```python
import math
```

```python
data_sample_size_sqrt=math.sqrt(35)
data_sample_size_sqrt
```

```python
standard_error = pop_std/data_sample_size_sqrt
standard_error
```

# Sampling distribution for the population of Klout

# n = sample size = 35
# group A's Klout score = 40 
# the population mean = 37.72

# What is z score of group A? (=what is z socre for a mean of 40?) 
# (=How many standard deviations above the mean of this distribution (distribution for the population of Klout) does this mean lie ? )



# ( group A socre - the population mean) / the standard deviation

```python
group_score=40
```

```python
(group_score-mean_pop)/standard_error
```
